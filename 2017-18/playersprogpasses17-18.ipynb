```python
# the famous import font code to use Andale Mono
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])
```

```python
#set default colors
text_color = 'white'
background = '#313332'
```

```python
#create a new column for progressive passes per 90
df['per90'] = df['Prog']/df['90s']
df
```

```python
#filter the dataframe so it is only players who have played more than 6.5 90's which is about 585 minutes
df = df[df['90s']>=6.5].reset_index()
df
```

```python
df = df.sort_values(by='per90',ascending=False)
```

```python
df = df[df['Pos'] != 'GK']
```

```python
df.head(100)
```

```python
#Look for Utd Players

df3 = df[df['Squad'] == 'Manchester Utd']
df3
```

```python
fig, ax = plt.subplots(figsize=(10,5))
fig.set_facecolor(background)
ax.patch.set_facecolor(background)

#set up our base layer
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.color'] = text_color

ax.grid(ls='dotted',lw=.5,color='lightgrey',axis='y',zorder=1)
spines = ['top','bottom','left','right']
for x in spines:
    if x in spines:
        ax.spines[x].set_visible(False)

sns.swarmplot(x='per90',data=df,color='white',zorder=1)

#plot Pogba
plt.scatter(x=9.497908,y=0,c='red',edgecolor='white',s=50,zorder=2)
plt.text(s='Pogba',x=9.497908,y=-.04,c='red')

#plot Matic
plt.scatter(x=5.780347,y=0,c='orange',edgecolor='white',s=50,zorder=2)
plt.text(s='Matic',x=5.780347,y=-.04,c='orange')

#plot Herrera
plt.scatter(x=5.603448,y=0,c='yellow',edgecolor='white',s=50,zorder=2)
plt.text(s='Herrera',x=4.803448,y=.08,c='yellow')

plt.title('Progressive Passes for Man Utd players in the Premier League 2017/18',c=text_color,fontsize=14)

plt.xlabel('Progressive Passes per 90',c=text_color)
```
