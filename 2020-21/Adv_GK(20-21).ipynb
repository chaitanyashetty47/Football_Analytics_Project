```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table = pd.read_csv("C:\\Users\\Amin\\Documents\\manutd(2012-2021)\\UtdAdvGK(20-21).csv")
table.head()
```

```python
x=table['#OPA/90'].mean()  #avg of defensive actions outside box by GKs in 20/21
```

```python
y=table['/90'].mean()      #avg shot stopping of GKs in 20/21
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 10)


plt.plot(table['#OPA/90'],table['/90'],"*")
plt.axvline(x, ls = '--', color = 'orange')
plt.axhline(y, ls = '--', color = 'orange')

ax.set_title("PL Goalkeeping Actions (20/21)",size="25")
ax.set_ylabel("Post-Shot xG - Goals Allowed(GA) p90")
ax.set_xlabel("Defensive Action outside Penalty Area p90 (Sweeping Actions)             Data:fbref via Statsbomb")

#Quadrant Labels
ax.text(0.2,0.18,"Great Shot Stopper\nPoor Sweeper",color="green",size="13")
ax.text(0.2,-0.28,"Poor Shot Stopper\nPoor Sweeper",color="green",size="13")
ax.text(0.96,0.18,"Great Shot Stopper\nGreat Sweeper",color="green",size="13")
ax.text(0.96,-0.28,"Poor Shot Stopper\nGreat Sweeper",color="green",size="13")


#plot label for points
for i in range(len(table)):
    if table['Player'][i] == 'David de Gea':
        plt.text(table['#OPA/90'][i] + .01 , table['/90'][i] -.01, table['Player'][i], color='maroon' , size="12")
    elif table['Player'][i] == 'Dean Henderson':
        plt.text(table['#OPA/90'][i] + .01 , table['/90'][i] -.01, table['Player'][i], color='maroon' , size="12")
    elif table['Player'][i]=='Sam Johnstone':
        plt.text(table['#OPA/90'][i] + .008, table['/90'][i] +.002 , table['Player'][i])        
    elif table['Player'][i]=='Łukasz Fabiański':
        plt.text(table['#OPA/90'][i] - .01, table['/90'][i] +.005 , table['Player'][i])
    elif table['Player'][i]=='Alisson':
        plt.text(table['#OPA/90'][i] - .025 , table['/90'][i] -.015 , table['Player'][i])
    elif table['Player'][i]=='Nick Pope':
        plt.text(table['#OPA/90'][i] - .025 , table['/90'][i] -.015 , table['Player'][i]) 
    else:
        plt.text(table['#OPA/90'][i] +.01, table['/90'][i] -.01 , table['Player'][i] )
        

plt.show()
```
