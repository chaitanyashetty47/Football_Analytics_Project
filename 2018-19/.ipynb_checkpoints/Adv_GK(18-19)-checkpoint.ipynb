```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table = pd.read_csv("C:\\Users\\Amin\\Documents\\manutd(2012-2021)\\UtdAdvGK(18-19).csv")
table.head()
```

```python
x=table['#OPA/90'].mean()  #avg of defensive actions outside box by GKs in 18/19
```

```python
y=table['/90'].mean()      #avg shot stopping of GKs in 18/19
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 10)


plt.plot(table['#OPA/90'],table['/90'],"*")
plt.axvline(x, ls = '--', color = 'orange')
plt.axhline(y, ls = '--', color = 'orange')

ax.set_title("PL Goalkeeping Actions (18/19)",size="25")
ax.set_ylabel("Post-Shot xG - Goals Allowed(GA) p90")
ax.set_xlabel("Defensive Action outside Penalty Area p90 (Sweeping Actions)")

#Quadrant Labels
ax.text(0.28,0.2,"Great Shot Stopper\nPoor Sweeper",color="green",size="13")
ax.text(0.28,-0.28,"Poor Shot Stopper\nPoor Sweeper",color="green",size="13")
ax.text(0.96,0.2,"Great Shot Stopper\nGreat Sweeper",color="green",size="13")
ax.text(0.96,-0.28,"Poor Shot Stopper\nGreat Sweeper",color="green",size="13")

#plot label for points
for i in range(len(table)):
    if table['Player'][i] == 'David de Gea':
        plt.text(table['#OPA/90'][i] + .01 , table['/90'][i] -.01, table['Player'][i], color='maroon' , size="12")
    elif table['Player'][i]=='Kasper Schmeichel':
        plt.text(table['#OPA/90'][i] - .2 , table['/90'][i] -.01 , table['Player'][i])
    elif table['Player'][i]=='Artur Boruc':
        plt.text(table['#OPA/90'][i] - .13 , table['/90'][i] , table['Player'][i])
    elif table['Player'][i]=='Wayne Hennessey':
        plt.text(table['#OPA/90'][i] + .01 , table['/90'][i] , table['Player'][i]) 
    elif table['Player'][i]=='Jonas Lössl':
        plt.text(table['#OPA/90'][i] - .08 , table['/90'][i] +.01 , table['Player'][i]) 
    else:
        plt.text(table['#OPA/90'][i] +.01, table['/90'][i] -.01 , table['Player'][i] )
        

plt.show()
```
