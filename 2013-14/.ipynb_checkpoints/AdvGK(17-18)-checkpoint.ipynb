```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table = pd.read_csv("C:\\Users\\Amin\\Documents\\manutd(2012-2021)\\UtdAdvGK(17-18).csv")
table.head()
```

```python
x=table['#OPA/90'].mean()  #avg of defensive actions outside box by GKs in 17/18
```

```python
y=table['/90'].mean()      #avg shot stopping of GKs in 17/18
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 10)


plt.plot(table['#OPA/90'],table['/90'],"*")
plt.axvline(x, ls = '--', color = 'orange')
plt.axhline(y, ls = '--', color = 'orange')

ax.set_title("PL Goalkeeping Actions (17/18)",size="25")
ax.set_ylabel("Post-Shot xG - Goals Allowed(GA) p90")
ax.set_xlabel("Defensive Action outside Penalty Area p90 (Sweeping Actions)")

ax.text(0.28,0.28,"Great Shot Stopper\nPoor Sweeper",color="green",size="13")
ax.text(0.28,-0.4,"Poor Shot Stopper\nPoor Sweeper",color="green",size="13")
ax.text(1.1,0.28,"Great Shot Stopper\nGreat Sweeper",color="green",size="13")
ax.text(1.1,-0.4,"Poor Shot Stopper\nGreat Sweeper",color="green",size="13")

#plot label for points
for i in range(len(table)):    
    if table['Player'][i] == 'David de Gea':
        plt.text(table['#OPA/90'][i] + .01 , table['/90'][i] -.01 , table['Player'][i], color='maroon' , size="12")
    elif table['Player'][i] == 'Jack Butland':
        plt.text(table['#OPA/90'][i] + .01 , table['/90'][i], table['Player'][i])
    elif table['Player'][i] == 'Petr Čech':
        plt.text(table['#OPA/90'][i] + .01 , table['/90'][i], table['Player'][i])
    elif table['Player'][i] == 'Orestis Karnezis':
        plt.text(table['#OPA/90'][i] + .01 , table['/90'][i], table['Player'][i])
    else:
        plt.text(table['#OPA/90'][i] +.01, table['/90'][i] -.01 , table['Player'][i] )

plt.show()
```
