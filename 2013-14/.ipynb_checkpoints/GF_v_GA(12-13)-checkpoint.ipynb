```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table = pd.read_csv("C:\\Users\\Amin\\Documents\\manutd(2012-2021)\\GF_v_GA(12-13).csv")
table.head()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 10)

plt.plot(table['GF'],table['GA'],"*")
plt.axvline(table['GF'].mean(), ls = '--', color = 'orange')
plt.axhline(table['GA'].mean(), ls = '--', color = 'orange')

ax.set_title("Premier League GF vs GA (12/13)",size="25")
ax.set_xlabel("Goals Scored(GF)")
ax.set_ylabel("Goals Conceded(GA)")

#plot label for points
for i in range(len(table)):
    if table['Squad'][i] == 'Manchester Utd':
        plt.text(table['GF'][i] - 8.7, table['GA'][i] + .8 , table['Squad'][i], color='maroon' , size="13")
    elif table['Squad'][i] == 'Southampton':
        plt.text(table['GF'][i] - 8, table['GA'][i] , table['Squad'][i])
    elif table['Squad'][i] == 'Reading':
        plt.text(table['GF'][i] - 5, table['GA'][i] + .5 , table['Squad'][i]) 
    else:
        plt.text(table['GF'][i] + 1, table['GA'][i] , table['Squad'][i] )

ax.text(35,40,"Poor Attack\nStrong Defense",color="red",size="10")
ax.text(65,40,"Strong Attack\nStrong Defense",color="red",size="10")
ax.text(35,70,"Poor Attack\nPoor Defense",color="red",size="10")
ax.text(65,70,"Strong Attack\nPoor Defense",color="red",size="10")

plt.show()
```
