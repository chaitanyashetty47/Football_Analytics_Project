```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table = pd.read_csv("C:\\Users\\Amin\\Documents\\manutd(2012-2021)\\UtdGFvGA(13-14).csv")
table.head()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 10)

plt.plot(table['GF'],table['GA'],"*")
plt.axvline(table['GF'].mean(), ls = '--', color = 'orange')
plt.axhline(table['GA'].mean(), ls = '--', color = 'orange')

ax.set_title("GF vs GA (13/14)",size="25")
ax.set_xlabel("Goals Scored(GF)")
ax.set_ylabel("Goals Conceded(GA)")

#plot label for points
for i in range(len(table)):
    if table['Squad'][i] == 'Manchester Utd':
        plt.text(table['GF'][i] + 1, table['GA'][i] , table['Squad'][i], color='maroon' , size="12")
    elif table['Squad'][i] == 'Manchester City':
        plt.text(table['GF'][i] - 9, table['GA'][i] + 1 , table['Squad'][i])
    elif table['Squad'][i] == 'Liverpool':
        plt.text(table['GF'][i] - 2.5, table['GA'][i] + 1, table['Squad'][i])
    elif table['Squad'][i] == 'West Brom':
        plt.text(table['GF'][i] - 11, table['GA'][i], table['Squad'][i])
    else:
        plt.text(table['GF'][i] + 1, table['GA'][i] , table['Squad'][i] )

ax.text(40,35,"Poor Attack\nStrong Defense",color="red",size="10")
ax.text(85,35,"Strong Attack\nStrong Defense",color="red",size="10")
ax.text(40,80,"Poor Attack\nPoor Defense",color="red",size="10")
ax.text(85,80,"Strong Attack\nPoor Defense",color="red",size="10")

plt.show()
```
