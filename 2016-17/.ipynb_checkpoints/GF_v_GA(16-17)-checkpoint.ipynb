```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table = pd.read_csv("C:\\Users\\Amin\\Documents\\manutd(2012-2021)\\UtdGFvGA(16-17).csv")
table.head()
```

```python
table['GA'].mean()
```

```python
table['GF'].mean()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(16, 12)

plt.plot(table['GF'],table['GA'],"*")
#plt.plot([table['GA'].mean(),table['GA'].mean()],[90,10],'k-', linestyle = ":", lw=2)
#plt.plot([5,115.4],[table['GF'].mean(),table['GF'].mean()],'k-', linestyle = ":", lw=2)
plt.axvline(60, ls = '--', color = 'orange')
plt.axhline(45, ls = '--', color = 'orange')

ax.set_title("GF vs GA (16-17)",size="25")
ax.set_xlabel("Goals Scored(GF)")
ax.set_ylabel("Goals Conceded(GA)")

#plot label for points
for i in range(len(table)):
    if table['Squad'][i] == 'Manchester Utd':
        plt.text(table['GF'][i] + 1, table['GA'][i] , table['Squad'][i], color='maroon' , size="12")
        
    elif table['Squad'][i] == 'Leicester City':
        plt.text(table['GF'][i]- 3, table['GA'][i] - 1, table['Squad'][i])    
    else:
        plt.text(table['GF'][i] +1, table['GA'][i] , table['Squad'][i] )

ax.text(40,40,"Poor Attack\nStrong Defense",color="red",size="10")
ax.text(65,38,"Strong Attack\nStrong Defense",color="red",size="10")
ax.text(40,70,"Poor Attack\nPoor Defense",color="red",size="10")
ax.text(65,70,"Strong Attack\nPoor Defense",color="red",size="10")

plt.show()
```
